```python
import pandas as pd
import numpy as np
```

```python
nama_file_excel = '/content/Data Coba.xlsx'
nama_file_csv_output = '/content/Data Coba_diperbarui.csv'

try:
    df_asli = pd.read_excel(nama_file_excel)
    num_rows = len(df_asli)

    first_names = ['Budi', 'Ani', 'Joko', 'Dewi', 'Sri', 'Agus', 'Putu', 'Made', 'Nyoman', 'Ketut']
    last_names = ['Santoso', 'Wijaya', 'Kusumo', 'Lestari', 'Setiawan', 'Tan', 'Wicaksono', 'Pratama']

    npp_komp_list = [f'{20250000 + i}' for i in range(1, num_rows + 1)]
    nama_list = [f'{np.random.choice(first_names)} {np.random.choice(last_names)}' for _ in range(num_rows)]

    df_asli['Npp_komp'] = npp_komp_list
    df_asli['Nama'] = nama_list

    df_asli.to_csv(nama_file_csv_output, index=False)

    print(f"File Excel telah berhasil diproses dan disimpan sebagai file CSV baru: '{nama_file_csv_output}'")
    print("\nBerikut adalah 5 baris pertama dari data yang sudah diperbarui:")
    print(df_asli.head())

except FileNotFoundError:
    print(f"Error: File '{nama_file_excel}' tidak ditemukan. Pastikan nama file dan lokasinya sudah benar.")
except Exception as e:
    print(f"Terjadi error: {e}")
```
